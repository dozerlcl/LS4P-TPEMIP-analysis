```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd
```

## 数据读入
前处理部分 merge_multfiles.ipynb

```python
dir_in ="/raid61/LS4P-TPEMIP/REFERENCE/CN05.1/"
ds_obs = xr.open_dataset(dir_in + "sel_CN05.1_Pre_1961_2018_daily_025x025.nc")
dir_in ="/raid61/LS4P-TPEMIP/TEMP_DATA/precip/"
ds_sys = xr.open_dataset(dir_in + "SYS_pre.nc").rename({"Time":"time"})
ds_nju_B = xr.open_dataset(dir_in + "NJU_LS4P_B_pre.nc")
ds_nju_C = xr.open_dataset(dir_in + "NJU_LS4P_C_pre.nc")
ds_yzx    = xr.open_dataset(dir_in + "Yang-ZhouX_pre.nc").rename({"Time":"time"})
ds_ysu    = xr.open_dataset(dir_in + "YSU_pre.nc")
ds_jamstecwrf    = xr.open_dataset(dir_in + "JAMSTECWRF_pre.nc")
ds_cwrf    = xr.open_dataset(dir_in + "CWRF_pre.nc")
# 整理到字典 mod_var 
var = {}
var['obs'] = ds_obs['pre']
var['sys'] = ds_sys['dailypre']
var["nju_B"] = ds_nju_B['pr']
var["nju_C"] = ds_nju_C['pr']
var["yzx"] = ds_yzx['dailypre']
var['ysu'] = ds_ysu['pr']
var['jam'] = ds_jamstecwrf['pr']
var['cwrf'] = ds_cwrf['pr']
```

## 平均态计算

```python
var_am = {}
var_jja= {}
for single_mod in var:
    var_am[single_mod] = var[single_mod].sel(time = var[single_mod].time.dt.month.isin([4,5])).mean('time') 
    var_jja[single_mod] = var[single_mod].sel(time = var[single_mod].time.dt.month.isin([6,7,8])).mean('time')
```
